```python
import pandas as pd
import numpy as np
import seaborn as sns

sns.set_style("darkgrid")
```

```python
df = pd.read_excel("wpa.xlsx")
df.head()
```

```python
df['last_value'] = df.value.shift(1)
df['last_match'] = df.match.shift(1)
df['wpa'] = np.where(df.last_match == df.match, df.value - df.last_value, 0)
df['wpa_abs'] = np.absolute(df.wpa)
```

```python
df_agg = df.groupby("match").agg(["mean", "sum"])[[("wpa_abs", "mean"),("wpa_abs", "sum")]]
df_agg.columns = ["wpa_mean","wpa_sum"]
df_agg.sort_values("wpa_sum",ascending=False)
```
